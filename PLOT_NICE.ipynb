```python
x = np.linspace(-1,1,100)   
y = -x

fig = plt.figure(figsize=(7,5)) 
ax = fig.gca()
ax.tick_params(axis='both', which='major', labelsize=16)
#ax.tick_params(axis='both', which='both', direction='in')
ax.set_xlabel(r'$x$', fontsize=21)
ax.set_ylabel(r'$y$', fontsize=21)

plt.plot(x,y,color='red', linewidth=2)
ax.set_title(r'${\rm Integration\,\, region}$', fontsize=20, color='black')   
ax.grid(linewidth=0.5)
ax.axvline(0, ymin=0, ymax=1, color='tab:blue',  linewidth=2., ls='dashed')
```

```python
x = np.linspace(-1,1,100)   
y = x
y2 = (1-x)/2

sns.set()
fig = plt.figure(figsize=(7,5)) 
ax = fig.gca()
ax.tick_params(axis='both', which='major', labelsize=16)
#ax.tick_params(axis='both', which='both', direction='in')
ax.set_xlabel(r'$x$', fontsize=21)
ax.set_ylabel(r'$y$', fontsize=21)
ax.axhline(0, xmin=0, xmax=1, color='red',  linewidth=1.5, ls='dashed')
ax.set_xlim(-0.5, 1) 
ax.set_ylim(-0.5, 1) 

plt.plot(x,y,color='tab:orange', linewidth=2)
plt.plot(x,y2,color='tab:blue', linewidth=2)
ax.set_title(r'${\rm Integration\,\, region}$', fontsize=20, color='black')   
ax.grid(linewidth=0.5)
```
